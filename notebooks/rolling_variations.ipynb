```python
%matplotlib inline
%pylab inline
pylab.rcParams['figure.figsize'] = (200, 100)

import pandas as pd
```

```python
names = ['sell', 'buy', 'last', 'vol', 'high', 'low', 'date']

data = pd.read_csv('btc.csv', names=names, skiprows=1)
```

```python
short_window = 400
long_window = 1000
very_long_window = 3000

short_variation = data['sell'].rolling(window=short_window, min_periods=1, center=False).mean()
long_variation = data['sell'].rolling(window=long_window, min_periods=1, center=False).mean()
very_long_variation = data['sell'].rolling(window=very_long_window, min_periods=1, center=False).mean()
```

```python
data['sell'].plot(grid=True)
short_variation.plot(grid=True)
long_variation.plot(grid=True)
very_long_variation.plot(grid=True)
```
